# Operators in Python

```python
# There are 7 types of operators in python
# 1. Arithemetic operators
# 2. Comparison operators
# 3. Logival Operators
# 4. Assignemnt Operators
# 5. Identify Operators
# 6. Membership Operators
# 7. Bitwise Operators
```

```python
# 1. Arithemetic => "+, -, *, /, //, **, %"
print(3+4) # Add
print(3-4) # Subtract
print(3*4) # Multiply
print(3/4) # Divide
print(3//4) # floor Division
print(3**4) # Exponentiation
print(3%4) # Modulus
```

```python
# 2. Comparison operators => ">, <, >=, <=, ==, !="
print(a>b) # Greater Than
print(a<b) # Less Than
print(a>=b) # Greater than or equal to
print(a<=b) # Less than or equal to
print(a==b) # Equal to
print(a!=b) # Not equal to
```

```python
# 3. Logival Operators can be also called as boolean => "and, or, not"
print(a>b and b<a) # and
print(a>b or b<a) # or
print(not(a>b or b<a)) # not
```

```python
# 4. Assignemnt Operators => "=, +=, -=, *="
score = 3 # =
print(score)

score +=4 # +=
print(score)

score -=4 # -=
print(score)

score *=4 # *=
print(score)
```

```python
# 5. Identify Operators => "is, is not"

a = 1
b = 2
print(a is b)

a = 1
b = 2
print(a is not b)
```

```python
# 6. Membership Operators => "in, not in"

a = 1,2,3,4
print(5 in a)


a = 1,2,3,4
print(5 not in a)
```

# Conditional Statements in Python

```python
# Types of condition statements
#     1. If Statement
#     2. If-else statement
#     3. if-elif-else statement
#     4. Nested if statement
#     5. short hand if statement
#     6. short hand if else statement
```

```python
# 1. If Statement

marks = int(input("Enter your marks here: "))

if marks >44:
    print("pass")
```

```python
# 2. If-else statement
marks = int(input("Enter your marks here: "))

if marks >44:
    print("pass")
else:
    print("fail")
```

```python
# 3. if-elif-else statement
marks = int(input("Enter your marks here: "))

if marks >= 90:
    print("Excellent")
    if marks >= 95:
        print("Excellent work!")
elif marks >= 80:
    print("Great Work!")
elif marks >= 40:
    print("pass!")
else:
    print("fail")
```

```python
# 4. Nested if statement
username = "python"
password = "123"

user_name = input("Enter your username: ")
pass_word = input("Enter your password: ")

if user_name == username and pass_word == password:
    print("Login in sucessfull!")
elif user_name == username and pass_word != password:
    print("Wrong pass word!")
    print("Enter your passowrd again")
    pass_word = input("Enter your password: ")
    if pass_word == password:
        print("Finally login accepted!")
    else:
        print("Access denied 1!")
else:
    print("Access denied 2!")
```

```python
# 5. short hand if statement
marks = int(input("Enter your marks: "))
if marks > 40: print("pass")
```

```python
# 6. short hand if else statement

marks = int(input("Enter your marks: "))
print("pass") if marks >40 else print ("fail")

```

# Merge and join in Pandas

```python
import pandas as pd
```

```python
data1 = {"Emp_id": ["E01","E02","E03","E04","E05","E06"],
        "Names": ["A","B","C","D","E","F"],
        "Age": ["25","29","35","21","63","45"]}

data2 = {"Emp_id": ["E01","E02","E03","E04","E05","E06","E07","E08"],
         "Salary": ["25000","36000","45000","18000","55000","70000","80000","72000"]}

df1 = pd.DataFrame(data1)
df2= pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
#merge data => inner join only common data in both tables
print(pd.merge(left = df1, right = df2, on = "Emp_id"))
```

```python
#outer join
print(pd.merge(left = df1, right = df2, on = "Emp_id", how = "outer"))
```

```python
#right join
print(pd.merge(left = df1, right = df2, on = "Emp_id", how = "right"))
```

```python
#left join
print(pd.merge(left = df1, right = df2, on = "Emp_id", how = "left"))
```
